2. Waarom is het nodig om vooraf te bekijken tot welk meetniveau elke variabele behoort?
Antwoord:
Het bepalen van het meetniveau van elke variabele is een fundamentele en cruciale eerste stap in data-analyse om de volgende redenen:
Keuze van Statistische Maten: Het meetniveau bepaalt welke centrum- en spreidingsmaten zinvol zijn.
Nominaal: Alleen de modus (meest voorkomende categorie) is zinvol.
Ordinaal: Modus en mediaan zijn zinvol. Een gemiddelde is technisch onjuist.
Ratio/Interval: Gemiddelde, mediaan, modus, standaarddeviatie en variantie zijn allemaal zinvol.
Keuze van Visualisaties: Het meetniveau dicteert de meest geschikte visualisatietechniek.
Nominaal/Ordinaal: Staafdiagrammen (count plots) of taartdiagrammen.
Ratio/Interval: Histogrammen, boxplots, of scatter plots (in relatie tot een andere numerieke variabele).
Keuze van Statistische Toetsen: De juiste statistische toets om hypotheses te onderzoeken hangt direct af van het meetniveau van de betrokken variabelen (bijv. Chi-kwadraat voor twee nominale variabelen, t-toets/ANOVA voor een nominale en een ratio variabele).
Voorbereiding voor Machine Learning (Preprocessing): Machine learning-algoritmes vereisen numerieke input. Het meetniveau bepaalt hoe we een variabele moeten transformeren:
Nominaal: Vaak via One-Hot Encoding om te voorkomen dat het model een onjuiste rangorde leert.
Ordinaal: Vaak via Ordinal Encoding (of Label Encoding) waarbij de rangorde wordt behouden (bijv. MBO=1, HBO=2, WO=3).
Ratio/Interval: Vaak via schalen (bijv. Standaardisatie of Normalisatie) om te zorgen dat variabelen met een groot bereik het model niet domineren.
Kortom, zonder kennis van het meetniveau lopen we het risico onjuiste analyses uit te voeren, misleidende visualisaties te maken en machine learning-modellen incorrect te trainen, wat leidt tot ongeldige conclusies.

6. Om te toetsen of de numerieke verdelingen (co2_uitstoot_per_jaar_KG en km_auto_per_jaar) normaal verdeeld zijn, gebruiken we de Shapiro-Wilk test. Deze test is geschikt voor het controleren van normaliteit. De hypothesen zijn:
H₀ (Nulhypothese): De data is afkomstig uit een normaal verdeelde populatie.
Hₐ (Alternatieve hypothese): De data is niet afkomstig uit een normaal verdeelde populatie.
We hanteren een significantieniveau (α) van 0.05. Als de p-waarde kleiner is dan 0.05, verwerpen we de nulhypothese. Een normaliteitstoets is niet van toepassing op de categorische variabele autobezit.

```python
# Functie om de Shapiro-Wilk test uit te voeren en te interpreteren
def normality_test(data, name):
    stat, p_value = stats.shapiro(data)
    print(f"Shapiro-Wilk Test voor '{name}':")
    print(f"  - Statistic: {stat:.4f}")
    print(f"  - p-waarde: {p_value:.4f}")
    if p_value < 0.05:
        print("  - Conclusie: De data is NIET normaal verdeeld (p < 0.05).")
    else:
        print("  - Conclusie: We kunnen niet uitsluiten dat de data normaal verdeeld is (p >= 0.05).")
    print("-" * 30)

# Voer de test uit voor de twee numerieke variabelen
normality_test(df_clean['co2_uitstoot_per_jaar_KG'], 'Jaarlijkse CO₂-uitstoot')
normality_test(df_clean['km_auto_per_jaar'], 'Jaarlijkse Autokilometers')
```

Voor beide variabelen is de p-waarde extreem klein (veel kleiner dan 0.05). We verwerpen daarom voor beide de nulhypothese. Dit bevestigt statistisch wat we visueel al zagen: noch de jaarlijkse CO₂-uitstoot, noch de jaarlijkse autokilometers zijn normaal verdeeld. Ze zijn beide significant scheef.

Machine learning modellen vereisen numerieke input. We moeten onze variabelen transformeren op basis van hun meetniveau. We slaan het resultaat op in een nieuw dataframe df_transformed om de originele waarden te behouden.
Per meetniveau:
Nominaal (bijv. geslacht, stad, burgerlijke_staat):
Wat: One-Hot Encoding.
Waarom: Deze methode creëert voor elke categorie een nieuwe, binaire (0 of 1) kolom. Bijvoorbeeld, de kolom geslacht wordt opgesplitst in geslacht_man, geslacht_vrouw, etc. Dit voorkomt dat het model een onjuiste, hiërarchische relatie tussen de categorieën leert (bijv. dat 'vrouw' (2) 'groter' is dan 'man' (1)).
Ordinaal (bijv. opleiding_niveau, gebruik_carsharing):
Wat: Ordinal Encoding.
Waarom: Deze methode kent aan elke categorie een numerieke waarde toe die de inherente rangorde respecteert. Bijvoorbeeld, voor opleiding_niveau mappen we MBO naar 1, HBO naar 2 en WO naar 3. Het model kan zo de volgorde ('WO is hoger dan HBO') correct interpreteren.
Ratio/Interval (alle numerieke kolommen):
Wat: Standaardisatie (Standard Scaling).
Waarom: Deze techniek transformeert de data zodat deze een gemiddelde van 0 en een standaarddeviatie van 1 heeft. Dit is essentieel voor veel modellen (zoals lineaire regressie, SVM, logistische regressie) die gevoelig zijn voor de schaal van de inputvariabelen. Zonder schalen zouden variabelen met een groot bereik (zoals jaarsalaris_eu) de variabelen met een klein bereik (zoals huishouden_grootte) onterecht domineren in het model.

```python
from sklearn.preprocessing import StandardScaler, OrdinalEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Maak een kopie voor transformatie
df_transformed = df_clean.copy()

# Definieer de mappings voor ordinale variabelen
opleiding_mapping = ['MBO', 'HBO', 'WO']
carsharing_mapping = ['Nooit', 'Heel soms', 'Soms', 'Vaak', 'Altijd']

# Maak lijsten van kolommen per type
# We laten persoon_ID eruit, want dat is geen feature
numerieke_cols = df_clean.select_dtypes(include=np.number).columns.drop(['persoon_ID', 'co2_uitstoot_per_jaar_KG'])
nominale_cols = ['geslacht', 'burgerlijke_staat', 'stad', 'provincie', 'autobezit', 'fietsbezit', 'elektrisch_auto']
ordinale_cols = ['opleiding_niveau', 'gebruik_carsharing']

# Maak een preprocessor pipeline
# Dit is de meest robuuste manier om transformaties consistent toe te passen
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerieke_cols),
        ('ord', OrdinalEncoder(categories=[opleiding_mapping, carsharing_mapping]), ordinale_cols),
        ('nom', OneHotEncoder(handle_unknown='ignore'), nominale_cols)
    ],
    remainder='passthrough' # Laat de overige kolommen (target) ongemoeid
)

# Pas de preprocessor toe
# Opmerking: de uitvoer is een NumPy array. Voor inspectie zetten we het terug naar een DataFrame.
# In een echt ML-project zouden we dit direct in een Pipeline met een model gebruiken.

# Fit en transform de data
transformed_data = preprocessor.fit_transform(df_transformed)

# Om de getransformeerde data te bekijken, maken we er een dataframe van
# Haal de nieuwe kolomnamen op uit de transformer
new_cols = list(numerieke_cols) + \
           list(ordinale_cols) + \
           list(preprocessor.named_transformers_['nom'].get_feature_names_out(nominale_cols))

# Creëer het dataframe (zonder de 'remainder' kolommen voor nu)
df_transformed_final = pd.DataFrame(transformed_data[:, :len(new_cols)], columns=new_cols)

print("Gedaante van het getransformeerde dataframe:", df_transformed_final.shape)
print("Eerste paar rijen van het getransformeerde dataframe:")
df_transformed_final.head()
```

De getransformeerde variabelen zijn nu opgeslagen en klaar voor gebruik in Deel D. Voor de volgende vragen in Deel A gebruiken we nog het df_clean dataframe, omdat dit beter te interpreteren is.

10. Ja, er is sprake van multicollineariteit.
De correlatiematrix toont een perfecte negatieve correlatie (r = -1.00) tussen dagen_thuiswerken_per_week_pro_rato en dagen_kantoorwerken_per_week_pro_rato. Dit betekent dat de ene variabele een perfecte lineaire voorspeller is van de andere.

Waarom is dit een probleem?
Multicollineariteit is vooral een probleem voor lineaire modellen (zoals lineaire en logistische regressie). 

De redenen zijn:
Onbetrouwbare coëfficiënten: Het model kan niet goed bepalen welk van de gecorreleerde variabelen verantwoordelijk is voor de invloed op de target. Hierdoor worden de coëfficiënten (de 'bèta's' in de formule) instabiel en moeilijk te interpreteren. Een kleine verandering in de data kan leiden tot grote schommelingen in de coëfficiënten.
Verkeerde interpretatie van significantie: Het kan ertoe leiden dat een variabele die in werkelijkheid significant is, als niet-significant wordt beoordeeld (hoge p-waardes voor de coëfficiënten), omdat de verklaarde variantie wordt "gedeeld" met de andere gecorreleerde variabele.

Hoewel het de voorspellende kracht van het model als geheel niet per se drastisch hoeft te verminderen, maakt het de interpretatie van de individuele bijdrage van elke variabele onmogelijk.

Mijn oplossing:
De oplossing is eenvoudig: verwijder een van de twee perfect gecorreleerde variabelen. Omdat dagen_thuiswerken en dagen_kantoorwerken perfect redundant zijn, bevat de ene variabele alle informatie die de andere ook bevat. We kunnen bijvoorbeeld de kolom dagen_kantoorwerken_per_week_pro_rato uit onze dataset verwijderen voordat we het model trainen, zonder enig verlies van informatie.

We kunnen echter wel de waarschijnlijkheidsdichtheid (Probability Density Function, PDF) op dat punt berekenen. De PDF-waarde geeft aan hoe "waarschijnlijk" waarden in de buurt van dat punt zijn. Een hogere PDF-waarde betekent een hogere concentratie van data rond dat punt.
We gebruiken de berekende gemiddelde (mu) en standaarddeviatie (sigma) van onze co2_uitstoot_per_jaar_KG kolom om dit te berekenen.

15. We stellen de hypothesen op om het verschil in de gemiddelde CO2-uitstoot tussen autobezitters en niet-autobezitters te toetsen.
H₀ (Nulhypothese): Er is geen significant verschil in de gemiddelde jaarlijkse CO2-uitstoot tussen personen die een auto bezitten en personen die geen auto bezitten.
Wiskundig: μ_autobezitters = μ_geen_auto
Hₐ (Alternatieve hypothese): Er is wel een significant verschil in de gemiddelde jaarlijkse CO2-uitstoot tussen personen die een auto bezitten en personen die geen auto bezitten.
Wiskundig: μ_autobezitters ≠ μ_geen_auto

16. Bij het toetsen van hypothesen kunnen twee soorten fouten worden gemaakt:
Type I Fout (α):
Definitie: Het onterecht verwerpen van de nulhypothese, terwijl deze eigenlijk waar is.
In deze context: We concluderen dat er een significant verschil is in CO2-uitstoot tussen autobezitters en niet-autobezitters, terwijl er in werkelijkheid geen verschil is. Dit is een "vals positief".
Type II Fout (β):
Definitie: Het onterecht niet verwerpen van de nulhypothese, terwijl deze eigenlijk onjuist is.
In deze context: We concluderen dat er géén significant verschil is in CO2-uitstoot, terwijl autobezitters in werkelijkheid wel degelijk een andere gemiddelde uitstoot hebben. Dit is een "vals negatief".

De juiste statistische toets is de Independent Samples T-test.
Waarom?
We willen de gemiddelden van een numerieke variabele (co2_uitstoot_per_jaar_KG, ratio meetniveau) vergelijken.
We vergelijken deze gemiddelden tussen twee onafhankelijke groepen (personen in de groep 'ja' zijn niet dezelfde als in de groep 'nee').
Aannames van de T-toets:
De waarnemingen zijn onafhankelijk (voldaan).
De data in elke groep is (ongeveer) normaal verdeeld.
De varianties van de twee groepen zijn gelijk (homoscedasticiteit).
We hebben bij vraag 6 al vastgesteld dat de CO2-data niet normaal verdeeld is. Echter, de t-toets is redelijk robuust tegen schendingen van normaliteit, vooral bij grote steekproeven zoals de onze (>30 per groep).
We moeten wel de aanname van gelijke varianties controleren. Dit doen we met de Levene's test. Als de varianties significant verschillen, voeren we de Welch's T-test uit, die geen gelijke varianties vereist. De ttest_ind functie in scipy.stats kan dit automatisch doen door equal_var=False in te stellen.

```python
# Stap 1: Splits de data in twee groepen
co2_met_auto = df_clean[df_clean['autobezit'] == 'ja']['co2_uitstoot_per_jaar_KG']
co2_zonder_auto = df_clean[df_clean['autobezit'] == 'nee']['co2_uitstoot_per_jaar_KG']

# Stap 2: Voer Levene's test uit om gelijke varianties te controleren
levene_stat, levene_p = stats.levene(co2_met_auto, co2_zonder_auto)
print(f"Levene's Test for Equal Variances:")
print(f"  - Statistic: {levene_stat:.4f}")
print(f"  - p-waarde: {levene_p:.4f}")

# Interpreteer Levene's test
if levene_p < 0.05:
    print("  - Conclusie: De varianties zijn significant verschillend (p < 0.05). We gebruiken Welch's T-test (equal_var=False).\n")
    equal_variances = False
else:
    print("  - Conclusie: De varianties zijn niet significant verschillend. We gebruiken de standaard T-test (equal_var=True).\n")
    equal_variances = True

# Stap 3: Voer de Independent Samples T-test uit
t_stat, p_value_ttest = stats.ttest_ind(co2_met_auto, co2_zonder_auto, equal_var=equal_variances)

print("Independent Samples T-test:")
print(f"  - T-statistic: {t_stat:.4f}")
print(f"  - p-waarde: {p_value_ttest:.100f}")

# Toon ook de gemiddelden ter context
print(f"\nGemiddelde CO2-uitstoot MET auto: {co2_met_auto.mean():.2f} KG")
print(f"Gemiddelde CO2-uitstoot ZONDER auto: {co2_zonder_auto.mean():.2f} KG")
```

22. Om de prestaties van dit regressiemodel te toetsen, zou ik een combinatie van twee metrics gebruiken:
Root Mean Squared Error (RMSE):
Wat het is: De vierkantswortel van het gemiddelde van de gekwadrateerde verschillen tussen de voorspelde en de werkelijke waarden.
Waarom:
Interpreteerbaar: De RMSE wordt uitgedrukt in dezelfde eenheid als de afhankelijke variabele (in dit geval, KG CO₂). Een RMSE van 1000 betekent dat het model er gemiddeld 1000 KG naast zit.
Bestraft grote fouten: Doordat de fouten worden gekwadrateerd, wegen grote fouten zwaarder mee dan kleine fouten. Dit is vaak wenselijk, omdat je liever geen enorme missers in je voorspelling hebt.
R-squared (R²):
Wat het is: Het percentage van de variantie in de afhankelijke variabele dat verklaard kan worden door de onafhankelijke variabelen in het model.
Waarom:
Relatieve maat: Waar RMSE een absolute foutmaat is, geeft R² een relatieve indicatie van de "goodness-of-fit". Een R² van 0.85 betekent dat 85% van de spreiding in CO2-uitstoot door ons model wordt verklaard, wat een zeer goede prestatie is. Het geeft context aan de RMSE-waarde.
Mijn keuze: De primaire metric is RMSE vanwege de directe interpreteerbaarheid, aangevuld met R² om een genormaliseerde, contextuele maat van de modelprestatie te hebben. In ons geval is de RMSE ~808 KG en de R² ~0.94, wat aangeeft dat het model zeer nauwkeurig is.
